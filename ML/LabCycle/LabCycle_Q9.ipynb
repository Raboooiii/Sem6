```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.manifold import TSNE
```

```python
iris = load_iris()
X = iris.data
y = iris.target

print("Original Shape:", X.shape)

def evaluate_model(X_transformed, y, method_name):
    clf = SVC(kernel='linear')
    scores = cross_val_score(clf, X_transformed, y, cv=5)
    print(f"{method_name} Accuracy: {scores.mean():.4f}")
```

```python
pca_2 = PCA(n_components=2)
X_pca_2 = pca_2.fit_transform(X)

evaluate_model(X_pca_2, y, "PCA (2D)")


lda_2 = LDA(n_components=2)
X_lda_2 = lda_2.fit_transform(X, y)

evaluate_model(X_lda_2, y, "LDA (2D)")


tsne_2 = TSNE(n_components=2, random_state=42)
X_tsne_2 = tsne_2.fit_transform(X)

evaluate_model(X_tsne_2, y, "t-SNE (2D)")


svd_2 = TruncatedSVD(n_components=2)
X_svd_2 = svd_2.fit_transform(X)

evaluate_model(X_svd_2, y, "SVD (2D)")
```

```python
pca_3 = PCA(n_components=3)
X_pca_3 = pca_3.fit_transform(X)

evaluate_model(X_pca_3, y, "PCA (3D)")


tsne_3 = TSNE(n_components=3, random_state=42)
X_tsne_3 = tsne_3.fit_transform(X)

evaluate_model(X_tsne_3, y, "t-SNE (3D)")


svd_3 = TruncatedSVD(n_components=3)
X_svd_3 = svd_3.fit_transform(X)

evaluate_model(X_svd_3, y, "SVD (3D)")
```
