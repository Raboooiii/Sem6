```python
from google.colab import files
uploaded = files.upload()

import pandas as pd

df = pd.read_csv('district_level_total_Registered_msme.csv')
df.head()
```

```python
df['small'] = df['small'].fillna(0)
df['medium'] = df['medium'].fillna(0)
```

```python
observed = [
    df['micro'].sum(),
    df['small'].sum(),
    df['medium'].sum()
]

observed

import numpy as np

total = sum(observed)
expected = [total/3]*3

expected

from scipy.stats import chisquare

chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis: Distribution is not equal")
else:
    print("Fail to Reject Null Hypothesis: Distribution is equal")
```

```python
table = df.groupby('state_name')[['micro','small','medium']].sum()

table.head()

from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table)

print("Chi-square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)

alpha = 0.05

if p < alpha:
    print("Reject Null Hypothesis: Enterprise type depends on state")
else:
    print("Fail to Reject Null Hypothesis: Enterprise type is independent of state")
```

```python
expected_df = pd.DataFrame(expected,
                           index=table.index,
                           columns=table.columns)

expected_df.head()
```
